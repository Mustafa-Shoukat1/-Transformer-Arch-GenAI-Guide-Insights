  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Chain of Thought (CoT)</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    Chain of Thought (CoT) is a recent advancement in prompting methods that encourages Large Language Models (LLMs) to explain their reasoning. This method contrasts with standard prompting by not only seeking an answer but also requiring the model to explain its steps to arrive at that answer.
  </p>

  <blockquote style="font-size: 16px; line-height: 1.6; margin-top: 10px;">
    The image above shows a comparison between a few-shot standard prompt (left) and a chain of thought prompt (right). This illustrates the difference: while the standard approach directly seeks a solution, the CoT approach guides the LLM to unfold its reasoning, often leading to more accurate and interpretable results.
  </blockquote>

## **Model size:**
####  CoT seems to be more effective for larger LLMs with stronger reasoning capabilities. Smaller models might struggle to follow the prompts or generate their own reasoning chains.

<div style="margin: 20px; padding: 20px; border: 2px solid #4a90e2; background-color: #f0f7fa; border-radius: 10px; box-shadow: 0 0 15px rgba(0, 0, 0, 0.1);">
  
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Self-Consistency with Chain of Thought (CoT-SC)</h2>
  
  <p style="font-size: 16px; line-height: 1.6;">
    Self-Consistency with Chain of Thought (CoT-SC) ensures that the language model finds the correct answer when asked the same question multiple times. When repeatedly queried, the model aggregates responses using a voting system to determine the most frequent answer, thereby converging on a single correct response.
  </p>
  
  <blockquote style="font-size: 16px; line-height: 1.6; margin-top: 10px;">
    The figure above demonstrates the iterative process of Chain of Thought with Self-Consistency (CoT-SC), highlighting how multiple queries lead to consensus on a definitive answer through a voting mechanism.
  </blockquote>

  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Procedure</h2>
  
  <ol style="font-size: 16px; line-height: 1.6;">
    <li><strong>Augmented Question:</strong> Modify the original question by adding "think step-by-step".</li>
    <li><strong>Repeated Questioning:</strong> Ask the question multiple times (n times) and collect all responses.</li>
    <li><strong>Voting Technique:</strong> Use a voting system to select the final answer. The response with the highest occurrence among collected answers is chosen as correct.</li>
  </ol>
  
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Advantages / Performance Gains</h2>
  
  <ul style="font-size: 16px; line-height: 1.6;">
    <li><strong>Addresses Major Issues of Large Language Models (LLMs):</strong></li>
    <li><strong>Repetitiveness and Local-Optimality in Greedy Decoding:</strong> LLMs often favor the most probable next word, potentially missing broader context.</li>
    <li><strong>Stochasticity of a Single Sampled Generation:</strong> Randomness in token selection can increase diversity but lead to nonsensical responses.</li>
  </ul>
  
  <p style="font-size: 16px; line-height: 1.6;">
    This approach enhances the model’s ability to consistently provide accurate answers by synthesizing responses through a structured voting mechanism.
  </p>
  
</div>


<div style="margin: 20px; padding: 20px; border: 2px solid #4a90e2; background-color: #f0f7fa; border-radius: 10px; box-shadow: 0 0 15px rgba(0, 0, 0, 0.1);">
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Tree of Thoughts (ToT) Prompting Framework</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    The Tree of Thoughts (ToT) prompting framework employs a tree-like structure to guide the reasoning process. Each step in the reasoning path branches out into potential routes. At every branch, the language model evaluates whether continuing down that path will likely lead to a conclusive answer.
  </p>

  <p style="font-size: 16px; line-height: 1.6;">
    If the model determines that a particular branch of reasoning is unproductive or unlikely to yield a solution, it abandons that branch and explores other paths within the tree. This iterative process allows the model to explore multiple avenues simultaneously, discarding unpromising routes and focusing on those that show promise of reaching a satisfactory conclusion.
  </p>

  <p style="font-size: 16px; line-height: 1.6;">
    By dynamically adjusting its course based on ongoing evaluations, the ToT framework aims to navigate through complex reasoning tasks efficiently. Its goal is to arrive at coherent and insightful answers by continuously adapting its approach to the evolving context of the reasoning process.
  </p>

  <p style="font-size: 16px; line-height: 1.6;">
    The Tree of Thoughts (ToT) prompting framework utilizes a tree-like structure to guide the reasoning process. Each step in the reasoning path branches out into potential routes. The Language Model (LLM) is queried multiple times to generate potential answers for a set of training questions. These question-answer pairs are generated incrementally through intermediate steps.

  <p style="font-size: 16px; line-height: 1.6;">

<div style="margin: 20px; padding: 20px; border: 2px solid #4a90e2; background-color: #f0f7fa; border-radius: 10px; box-shadow: 0 0 15px rgba(0, 0, 0, 0.1);">
  
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">ReAct Prompting</h2>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - Yao et al., 2022 introduced a framework named ReAct where LLMs are used to generate both reasoning traces and task-specific actions in an interleaved manner.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - Generating reasoning traces allows the model to induce, track, and update action plans, and even handle exceptions. The action step allows the model to interface with and gather information from external sources such as knowledge bases or environments.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - The ReAct framework enables LLMs to interact with external tools to retrieve additional information, leading to more reliable and factual responses.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - Results show that ReAct can outperform several state-of-the-art baselines on language and decision-making tasks. ReAct also enhances human interpretability and trustworthiness of LLMs by integrating both internal knowledge and external information obtained during reasoning.
  </p>

</div>

<div style="margin: 20px; padding: 20px; border: 2px solid #4a90e2; background-color: #f0f7fa; border-radius: 10px; box-shadow: 0 0 15px rgba(0, 0, 0, 0.1);">
  
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">How Does ReAct Work?</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    - ReAct is inspired by the synergies between “acting” and “reasoning” which allow humans to learn new tasks and make decisions or reasoning.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - Chain-of-thought (CoT) prompting has shown the capabilities of LLMs to carry out reasoning traces to generate answers to questions involving arithmetic and commonsense reasoning, among other tasks (Wei et al., 2022). However, its lack of access to the external world or inability to update its knowledge can lead to issues like fact hallucination and error propagation.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - ReAct is a general paradigm that combines reasoning and acting with LLMs. ReAct prompts LLMs to generate verbal reasoning traces and actions for a task. This allows the system to perform dynamic reasoning to create, maintain, and adjust plans for acting while also enabling interaction with external environments (e.g., Wikipedia) to incorporate additional information into the reasoning.
  </p>

</div>

<div style="margin: 20px; padding: 20px; border: 2px solid #4a90e2; background-color: #f0f7fa; border-radius: 10px; box-shadow: 0 0 15px rgba(0, 0, 0, 0.1);">
  
  <h2 style="color: #4a90e2; font-family: 'Arial', sans-serif; text-shadow: 1px 1px 2px rgba(0,0,0,0.3); font-size: 24px; font-weight: bold; margin-bottom: 20px;">Reasoning WithOut Observation (ReWOO)</h2>

  <p style="font-size: 16px; line-height: 1.6;">
    - ReWOO (Reasoning WithOut Observation) is a technique that detaches the reasoning process from external observations, such as the ability to access and process information from the real world. This detachment significantly reduces the amount of tokens that the LLM needs to consume, which in turn improves the efficiency of the LLM.
  </p>
  
  <p style="font-size: 16px; line-height: 1.6;">
    - ReWOO divides the workflow into three modules: Planner, Worker, and Solver. The Planner breaks down a question into steps, each formulated as a plan. The Worker retrieves external knowledge, and the Solver synthesizes plans and evidence to generate answers.
  </p>
